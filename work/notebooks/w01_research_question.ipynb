## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Decision: Which declining pages should a content strategist review first,
given limited time to act on any of them?

Unit of analysis: one content page (content_id).

Action: The strategist gets a ranked list and works top-down, updating or
rewriting the highest-priority pages first.

Cost of a wrong call: If a stable page gets ranked high, the strategist wastes
review time on something that didn't need it. If a genuinely declining page
with real demand gets ranked low or missed, real traffic keeps bleeding away
unnoticed — a more expensive mistake, since it compounds over time.

Why data/ML helps: With 30,000 pages and dozens of signals (trend, position,
CTR, impressions, engagement) interacting at once, no one can rank this by
hand. A simple rule can flag "declining," but deciding WHICH declining pages
matter most needs a way to weigh several signals together — that's what a
score does that a single if-statement can't.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/nellypoghosyan/ML-FlyRank-AI/main/data/raw/content_refresh_anonymized.csv')

print("Total pages:", len(df))
print(df['trend_direction'].value_counts())

declining_with_demand = ((df['trend_direction']=='down') & (df['impressions_90d']>=100)).sum()
print("Declining pages with real demand (impressions_90d >= 100):", declining_with_demand)

low_ctr_visible = ((df['impressions_90d']>=500) & (df['avg_position']>0) &
                    (df['avg_position']<=20) & (df['ctr']<0.5)).sum()
print("Low-CTR visible pages (decent position, weak clicks):", low_ctr_visible)
```
